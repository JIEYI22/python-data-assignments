```python
import requests
import bs4
import csv
from bs4 import BeautifulSoup
```

```python
def get_the_page(url):  
    r=requests.get(url)
    mypage=BeautifulSoup(r.text)

    a = mypage.find('div',attrs={'class':'box searchList searchResult searchList-v1009 listModeView'})
    
    myname=a.find_all('li',attrs={'class':'shopname'})
    names=[]
    for i in myname:
        myname=i.find('a')
        names.append(myname.text)
 
    myprice=a.find_all('strong',attrs={'class':'average'})
    prices=[]
    for i in range(len(myprice)):
        prices.append(myprice[i].text)
 
    myaddress=a.find_all('li',attrs={'class':'address'})
    addresses=[]
    for i in range(0,15):
        addresses.append(myaddress[i].text.split('\n')[1])
        
    tels=[]
    for i in range(0,15):
        tels.append(myaddress[i].text.split('\n')[2])

    mytags=a.find_all('li',attrs={'class':'tags'})
    tags=[]
    for i in mytags:
        mytags=i.find('a')
        tags.append(mytags.text)
 
    data=[]
    for i in range (0,len(names)):
        data.append([names[i],prices[i],addresses[i],tels[i],tags[i]])
        
    return data
```

```python
data_list=[]
for i in range(0,20):
    page_number=str(i+1)
    url ='http://www.dianping.com/macau/food/p'+ page_number
    data_list.append(url)
```

```python
data_information=[]
for i in data_list:
    my=get_the_page(i)
    data_information.extend(my)
```

```python
header = ['Shopname','Average Price','Address','Tel','Tags']
import csv
with open('dianping22.csv', 'w', newline = '\n') as file:
	writer = csv.writer(file)
	writer.writerow(header)
	writer.writerows(data_information)
```
